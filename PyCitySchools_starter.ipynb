```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#total_schools = school_data_complete.groupby('school_name')['School ID'].nunique()
total_schools = school_data.groupby('school_name')['School ID'].nunique().count()
total_students = student_data.groupby('Student ID').nunique().count()
total_students = total_students[0]
total_budget =  school_data['budget'].sum()
total_math_score =  float(format(student_data['math_score'].mean(), '.2f'))
total_read_score =  float(format(student_data['reading_score'].mean(), '.2f'))
overal_passing_rate = float(format(((total_math_score + total_read_score) / 2), '.2f'))
students_math_70 = student_data[student_data.math_score >= 70].count()
students_math_70 = float(format(((students_math_70[0] * 100) / total_students), '.2f'))
students_read_70 = student_data[student_data.reading_score >= 70].count()
students_read_70 = float(format(((students_read_70[0] * 100) / total_students), '.2f'))
#df = df.groupby('domain')['ID'].nunique()
#print(f'Total Schools: {total_schools}')
#print(f'Total Students: {total_students}')
#print(f'Total Budget: {total_budget}')
#print(f'Math Score Average: {total_math_score}')
#print(f'Reading Score Average: {total_read_score}')
#print(f'Overal Passing Rate: {overal_passing_rate}')
#print(f'Students Math >= 70: {students_math_70}')
#print(f'Students Reading >= 70: {students_read_70}')

df_resumen = pd.DataFrame({
                'Total Schools' : [total_schools],
                'Total Students' : [total_students],
                'Total Budget' : [total_budget],
                'Math Score Average' : [total_math_score],
                'Reading Score Average' : [total_read_score],
                'Overal Passing Rate' : [overal_passing_rate],
                'Students Math >= 70' : [students_math_70],    
                'Students Reading >= 70' : [students_read_70]
})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#school_data_complete.columns
```

```python
#school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data.type
#df_tmp = pd.DataFrame(columns=['Sn', 'Ty'])
df_tmp = pd.DataFrame()
df_r = pd.DataFrame()
cols = ['School Name', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate']
for index, school in school_data.iterrows():
    sn = school['school_name']
    st = school['type']
        #student_data[student_data.math_score >= 70].count()
    ts = student_data[student_data.school_name == sn].count()[0]
    tsb = school['budget']
    tbps = school['budget'] / ts 
    ams = float(format(student_data[student_data.school_name == sn]['math_score'].mean(),'.2f'))
    ars = float(format(student_data[student_data.school_name == sn]['reading_score'].mean(),'.2f'))
    ppm = float(format(((student_data[(student_data.school_name == sn) & (student_data.math_score >= 70)].count()[0] * 100 ) / ts),'.2f'))
    ppr = float(format(((student_data[(student_data.school_name == sn) & (student_data.reading_score >= 70)].count()[0] * 100 ) / ts),'.2f'))
    opr = float(format(((ppm + ppr) / 2), '.2f'))
    #print(f'{sn} - {st} - {ts} - {tsb} - {tbps} - {ams} - {ars} - {ppm} - {ppr} - {opr}')
    
    df_tmp = [(sn, st, ts, tsb, tbps, ams, ars, ppm, ppr, opr)]
    df_r = df_r.append(pd.DataFrame(df_tmp, columns=cols, index=[index]))    
    
    
```

## Math Scores by Grade

```python
df_tmp = pd.DataFrame()
df_r = pd.DataFrame()
cols = ['School Name', '9th', '10th', '11th', '12th']
for index, school in school_data.iterrows():
    sn = school['school_name']
    rs9  = format(student_data[(student_data.school_name == sn) & (student_data.grade == '9th')]['math_score'].mean(), '.2f')
    rs10 = format(student_data[(student_data.school_name == sn) & (student_data.grade == '10th')]['math_score'].mean(), '.2f')
    rs11 = format(student_data[(student_data.school_name == sn) & (student_data.grade == '11th')]['math_score'].mean(), '.2f')
    rs12 = format(student_data[(student_data.school_name == sn) & (student_data.grade == '12th')]['math_score'].mean(), '.2f')
    #rs10 = float(format(((student_data[(student_data.school_name == sn) & (student_data.reading_score >= 70)].count()[0] * 100 ) / ts),'.2f'))     
    #print(f'{sn} - {rs9} - ')
    
    df_tmp = [(sn, rs9, rs10, rs11, rs12)]
    df_r = df_r.append(pd.DataFrame(df_tmp, columns=cols, index=[index]))  
    
df_r.head()    
```
